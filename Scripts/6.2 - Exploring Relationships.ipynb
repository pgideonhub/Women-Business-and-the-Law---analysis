### Correlations

```python
# Create a subset excluding the 'country_year', 'country', 'country_code', 'region', 'income_group', 'year' columns

sub = df[['wbl_index', 'mobility', 'workplace', 'pay','marriage', 'parenthood', 'business', 'assets', 'pension']]
```

```python
sub
```

```python
# Create a correlation matrix using the subset created above

sub.corr()
```

```python
# Create a correlation heatmap for the subset using matplotlib

plt.matshow(sub.corr())
plt.show()
```

```python
# Save figure
plt.matshow(sub.corr())
plt.savefig("out.png") 
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub.shape[1]), sub.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub.shape[1]), sub.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, ax = ax)
```

The coefficients in the correlation matrix heatmap describe the relationship between the variables. A coefficient of 0 means no relationship; 0.1 to 0.3 identifies a week relationship, 0.3 to 0.5 a moderate relationship and 0.5 to 1.0 a strong relationship. The same strength applies for negative coefficients, but in the opposite direction (ie. if one variable increases, the other decreases). Of all the coefficients above, 0.79 is the highest - between the wbl_index and the workplace indicator.So, although all of the above indicators contribute to the final wbl_score, 'workplace' has the highest correlation.

### Scatterplots

```python
# Create a scatterplot for the "wbl_index" and "marriage" columns in seaborn

sns.lmplot(x = 'marriage', y = 'wbl_index', data = df)
```

Although all of the indicators contribute to the overall wbl_index, the 'workplace' and 'marriage' indicatars have the highest correlations, meaning the higher these indicator scores, the higher the wbl_index in relation. For example, the 'pension' indicator may increase, but the wbl_index will not increase as much as it would it the 'workplace' or 'marriage' indicators increase.

```python
# Create wbl_index categories: Low, Medium & High
```

```python
df.loc[df['wbl_index'] < 40, 'wbl_category'] = 'Low index'
```

```python
df.loc[(df['wbl_index'] >= 40) & (df['wbl_index'] < 80), 'wbl_category'] = 'Middle index'
```

```python
# Check frequency of index categories
df['wbl_category'].value_counts(dropna = False)
```

```python
# Create a categorical plot in seaborn using the wbl categories created above with 'income_group'

sns.set(style="ticks")
g = sns.catplot(x="income_group", y="wbl_index", hue="wbl_category", data=df, height=8, aspect=1)
```
